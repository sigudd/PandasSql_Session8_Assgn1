Read the following data set: https://archive.ics.uci.edu/ml/machine-learning-databases/adult/ Rename the columns as per the description from this file: https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.names Task: Create a sql db from adult dataset and name it sqladb

Select 10 records from the adult sqladb
Show me the average hours per week of all men who are working in private sector
Show me the frequency table for education, occupation and relationship, separately
Are there any people who are married, working in private sector and having a masters degree
What is the average, minimum and maximum age group for people working in different sectors
Calculate age distribution by country
Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'

```python
import pandas as pd
import sqlite3
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns1 = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','range']
adults = pd.read_csv(url,header=None,names=columns1)
```

```python
con = sqlite3.connect("sqladb.db")
```

```python
con.execute("create table adult2('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','range')")
```

```python
con.executemany("insert into adult2('age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race','sex','capital_gain','capital_loss','hours_per_week','native_country','range') values (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)", adults.values.tolist())
```

```python
cursor = con.execute("SELECT * from adult2 limit 10")
for row in cursor:
   print(row)
```

```python
sql = "select avg(hours_per_week) from adult2 where trim(workclass)='Private' "
cursor = con.execute(sql)
for row in cursor:
    print ("Average Work Hours per Week for Private sector employees = ", row[0])
```

```python

sql = "select education, count (education) 'Frequency' from adult2 group by education"
cursor = con.execute(sql)
print("Eduction \t Frequency")
for row in cursor:
    print(row[0],'\t',row[1])
```

```python
sql = "select occupation, count (occupation) 'Frequency' from adult2 group by occupation"
cursor = con.execute(sql)
print("Occupation \t Frequency")
for row in cursor:
    print(row[0],'\t',row[1])
```

```python
sql = "select occupation, count (relationship) 'Frequency' from adult2 group by relationship"
cursor = con.execute(sql)
print("Relationship \t Frequency")
for row in cursor:
    print(row[0],'\t',row[1])
```

```python
sql = "select count(*) from adult2 where  trim(workclass)='Private' and trim(education)='Masters'  and trim(marital_status) like 'Married-%'" 
cursor = con.execute(sql)

for row in cursor:
    print('people who are married, working in private sector and having a masters :' ,row[0])
```

```python
sql = "select workclass,avg(age),min(age),max(age) from adult2 group by workclass" 
cursor = con.execute(sql)
print('workclass','\t','avg','\t\t', ' min ','\t', 'max')
for row in cursor:
    print(row[0],'\t',row[1],'\t',row[2],'\t',row[3])
    #print(row)
```

```python
sql = "select native_country,age,count(*) from adult2 group by native_country,age" 
cursor = con.execute(sql)
print('native country','age', 'age_count')
for row in cursor:
    print(row[0],'\t',row[1],'\t',row[2])
```

```python
sql = "select (capital_gain-capital_loss) as 'Net_Capital_Gain' from adult2 " 
cursor = con.execute(sql)
print('Net-Capital-Gain')
for row in cursor:
    print(row[0])
```
